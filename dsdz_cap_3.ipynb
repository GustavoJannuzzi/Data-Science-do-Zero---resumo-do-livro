## matplotlib
Neste capítulo usaremos matplotlib.pyplot.

### Gráfico de linha 

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# cria um gráfico de linha, anos no eixo x e gdp no y.
plt.plot(years, gdp, color='green')

# Título
plt.title("Nominal GDP")

# label(legenda) no eixo y 
plt.ylabel("Billions of $")
# plt.show()
```

### Gráficos de barra/ Histogramas

```python
# Este gráfico mostra quantas premiações do Oscar cada variedade ganhou.

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [num_oscars]
plt.bar(movies, num_oscars)

plt.title("Meu filme favorito")     # add a title
plt.ylabel("# do Academy Awards")   # label the y-axis

# label x-axis with movie names at bar centers
plt.xticks(movies)

plt.show()
```

```python
# Histogram

from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

plt.bar([x + 5 for x in histogram.keys()],  # Shift bars right by 5
        histogram.values(),                 # Give each bar its correct height
        8,                                 # Give each bar a width of 8
        edgecolor=(0, 0, 0))                # Black edges for each bar

plt.axis([-5, 105, 0, 5])                  # eixo x de -5 a 105 // eixo y de 0 a 5
                                           

plt.xticks([10 * i for i in range(11)])    # x-axis labels at 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

```python
mentions = [500, 505]
years = [2013, 2014]

plt.bar([2012.6, 2013.6], mentions, 0.8)
plt.xticks(years)
plt.ylabel("# de vezes que ouvimos alguém dizer 'Data Science'")

```

### Gáficos de Linhas

```python
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

#Podemos faze múltiplas chamadas em plt.plot
# assim mostra múltilpas séries no mesmo gráfico
plt.plot(xs, variance, 'g-', label = 'Variance')       #linha verde sólida
plt.plot(xs, bias_squared, 'r-.', label ='bias^2')     #linha de ponto tracejado vermelho
plt.plot(xs, total_error, 'b:', label = 'total error') #linha azul pontilhada

#Atribuir multiplas rótulos para cada série
# podems abter uma legenda gratuita
#loc = 9 significa "top center"
plt.legend(loc=9)
plt.xlabel("complexidade do modelo")
plt.title("Compromisso entre polarização e variância")
plt.show()
```

### Gráficos de Disperção
Escolha certa pra visualizar o relacionamento de dois pares de conjuntos de dados. <br>
No exemplo abaixo o gráfico mostra a relação entre o número de amigos que os usuários tem e o número de minutos que eles passam no site por dia.

```python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

#nomeia cada posição 
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                xy=(friend_count, minute_count),   #coloca o rótulo com sua posição
                xytext = (5, -5),                  #mas compensa um pouco
                textcoords = 'offset points')
plt.title("Minutos Diários vs. Número de Amigos")
plt.xlabel("# de amigos")
plt.ylabel("minutos diários no sites")
plt.show()
```

Se estivermos espalhando variáveis comparáveis, talvez tenhamos uma imagem enganosa, como apresentado abaixo. Para isso não acontecer, ajustamos a escala na proxima imagem.

```python
# Imagem muito espalhada
test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Os eixos não são cmpatíveis")
plt.xlabel("nota do teste 2")
plt.ylabel("nota do teste 1")
plt.show()
```

```python
# Imagem ajustada com axis("equal")
plt.scatter(test_1_grades, test_2_grades)
plt.axis("equal")
plt.title("Os eixos não são cmpatíveis")
plt.xlabel("nota do teste 2")
plt.ylabel("nota do teste 1")
plt.show()
```
